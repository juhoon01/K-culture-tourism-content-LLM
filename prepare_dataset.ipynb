```python
# 데이터 정제
# 우리가 학습에 필요한 변수는 raw_dataset/"한국어", Label/"question"과 "answer" 
# 한국어는 결측치 있어서 제거해야함, 질문 대답 변수는 결측치 없음
import pandas as pd

Food_data = pd.read_csv("Training/raw_dataset/FOOD_all.csv")
HIS_data = pd.read_csv("Training/raw_dataset/HIS_all.csv")
LEI_data = pd.read_csv("Training/raw_dataset/LEI_all.csv")
NAT_data = pd.read_csv("Training/raw_dataset/NAT_all.csv")
SHOP_data = pd.read_csv("Training/raw_dataset/SHOP_all.csv")
STAY_data = pd.read_csv("Training/raw_dataset/STAY_all.csv")

raw_df = [Food_data,HIS_data,LEI_data,NAT_data,SHOP_data,STAY_data]
#원천 데이터 다 합치기
Raw_dataset = pd.concat(raw_df,ignore_index=True)
# 인덱스 다시 설정
Raw_dataset.reset_index(drop=True, inplace=True)

print(Raw_dataset.shape)
Raw_dataset.head()
```

```python
# 결측치 찾아서 제거

# (1) 한국에 데이터 중 결측치를 가진 데이터 제거하기
# NAN값인 데이터만 확인 - 12000개정도있음
null_rows =Raw_dataset[Raw_dataset["한국어"].isnull()]
null_rows.head()
# '한국어' 컬럼에서 결측치 행 제거
cleaned_df = Raw_dataset.dropna(subset=["한국어"])
# 인덱스 재정렬
cleaned_df = cleaned_df.reset_index(drop=True)
print(f"원본 행 수: {len(Raw_dataset)}, 결측치 제거 후 행 수: {len(cleaned_df)}")
```

```python
# 원본 데이터셋 다시 저장
cleaned_df.to_csv("raw_dataset.csv",index=False,encoding="utf-8-sig")
```

```python
import glob
import json
import pandas as pd
# (2)웝본 데이터에 대한 질문, 답변 데이터 합치기
# glob으로 모든 JSON 파일 경로 수집
patterns = [
    "Training/label_dataset/FOOD/*.json",
    "Training/label_dataset/HIS/*.json",
    "Training/label_dataset/LEI/*.json",
    "Training/label_dataset/NAT/*.json",
    "Training/label_dataset/SHOP/*.json",
    "Training/label_dataset/STAY/*.json",
]
all_records = [] 
for pattern in patterns:
    # 1) 폴더 내 파일 오름차순 정렬
    json_paths = sorted(glob.glob(pattern))
    for fp in json_paths:
        with open(fp, 'r', encoding='utf-8') as f:
            data = json.load(f)
        # 2) JSON이 리스트면 리스트 확장, 단일 객체면 append
        if isinstance(data, list):
            all_records.extend(data)
        else:
            all_records.append(data)
QA_df = pd.DataFrame(all_records)
print(QA_df.shape)
QA_df.head()
```

```python
# 순서 잘 정렬된거 확인
QA_df['text'][0] == Raw_dataset["일본어"][0]
```

```python
# data_info 컬럼에서 documentID를 꺼내서 새로운 컬럼 생성 - QA랑 한국어 원본이랑 매칭하기 위해서
QA_df['documentID'] = QA_df['data_info'].apply(lambda info: info['documentID'])

# QA 리스트 하나당 한 행이 되도록 
QA_df = QA_df.explode('QA').reset_index(drop=True)

# QA 컬럼에서 question, answer 꺼내서 별도 컬럼 생성 - JSONL 형식 구성할때 편하게 생성하기 위해 별도의 열로 나눔
QA_df['question'] = QA_df['QA'].apply(lambda qa: qa['question'])
QA_df['answer']   = QA_df['QA'].apply(lambda qa: qa['answer'])
# 더 이상 필요 없으면 QA 컬럼 제거
QA_df = QA_df.drop(columns=['QA'])

# 확인
print(QA_df.columns)

# (3) 결측치 제거된 한국어 데이터프레임이랑 QA 데이터랑 병합해서 학습 데이터 생성
train_df = pd.merge(
    QA_df,
    cleaned_df[['문단 번호','한국어']],
    left_on='documentID',
    right_on='문단 번호',
    how='inner'
)
# 병합 후 디버깅 - 결측치 제거된거 잘 반영됐는지 확인
print("병합 후 개수:", train_df.shape)
# 학습을 위해 JSON형식으로 저장
with open('QA_dataset.jsonl', 'w', encoding='utf-8') as fout:
    for _, row in train_df.iterrows():
        context  = row['한국어'].replace('\n', ' ').strip()
        question = row['question'].replace('\n', ' ').strip()
        answer   = row['answer'].strip()

        rec = {
            "instruction": "한국어 글을 읽고 질문에 답하세요",
            "input":       f"【本文（한국어）】\n{context}\n\n【質問（日本語）】\n{question}",
            "output":      answer
        }
        fout.write(json.dumps(rec, ensure_ascii=False) + "\n")
```

```python
from pathlib import Path

# 최종 학습 데이터 확인
jsonl_path = Path("QA_dataset.jsonl")

# 줄 수 확인
with jsonl_path.open(encoding="utf-8") as f:
    num_lines = sum(1 for _ in f)
print(f"총 레코드 수: {num_lines:,}")

# 5줄 확인
N = 5
with jsonl_path.open(encoding="utf-8") as f:
    for i, line in zip(range(N), f):
        print(line.strip()[:250], "…") 
        
# 3) key 구조 + input 내부 분해 확인 ────────────────
records = []
with jsonl_path.open(encoding="utf-8") as f:
    for _ in range(N):
        records.append(json.loads(next(f)))
# 확인용!! instruction/input/output 형태는 유지되어있음 질문이 원본과 잘 결합되어있나 확인하기 위함
print("\n=== PARSED SAMPLE (본문·질문 분리)")
for r in records:
    # input을 줄 단위로 분리
    lines = [ln.strip() for ln in r["input"].splitlines() if ln.strip()]
    # 한국어 본문은 두 번째 줄, 질문은 마지막 줄 가정
    context  = lines[1][:60] if len(lines) > 1 else ""
    question = lines[-1][:60] if lines else ""
    print({
        "instruction": r["instruction"],
        "context":     context + ("…" if len(lines[1]) > 60 else ""),
        "question":    question + ("…" if len(lines[-1]) > 60 else ""),
        "output":      r["output"][:60] + ("…" if len(r["output"]) > 60 else "")
    })
    
# pandas형태로도 확인
df = pd.read_json(jsonl_path, lines=True)
print(df.info())
df.head()
```

```python
# Val_dataset 전처리
import pandas as pd
# 1.엑셀 파일 읽어 합치기
xlsx_paths = [
    "Validation/raw_dataset/VS_FOOD/FOOD (113331~127498).xlsx",
    "Validation/raw_dataset/VS_HIS/HIS (115947~130439).xlsx",
    "Validation/raw_dataset/VS_LEI/LEI(14569~16389).xlsx",
    "Validation/raw_dataset/VS_NAT/NAT(36812~41412).xlsx",
    "Validation/raw_dataset/VS_SHOP/SHOP(29156~32799).xlsx",
    "Validation/raw_dataset/VS_STAY/STAY(62146~69913).xlsx",
]
dfs = []
for fp in xlsx_paths:
    df = pd.read_excel(fp, engine="openpyxl")   # Excel 전용 함수
    dfs.append(df)
raw_val_df = pd.concat(dfs, ignore_index=True).reset_index(drop=True)
print("원본 검증 원천 shape :", raw_val_df.shape)
raw_val_df.head()
```

```python
import glob
import json
import pandas as pd
# 웝본 데이터에 대한 질문, 답변 데이터 합치기
# glob으로 모든 JSON 파일 경로 수집
val_QA = [
    "Validation/label_dataset/VL_FOOD/*.json",
    "Validation/label_dataset/VL_HIS/*.json",
    "Validation/label_dataset/VL_LEI/*.json",
    "Validation/label_dataset/VL_NAT/*.json",
    "Validation/label_dataset/VL_SHOP/*.json",
    "Validation/label_dataset/VL_STAY/*.json",
]
val_records = []

for pattern in val_QA:
    # 해당 패턴에 매칭되는 파일 경로를 정렬해서 가져오기
    json_paths = sorted(glob.glob(pattern))
    for fp in json_paths:
        with open(fp, 'r', encoding='utf-8') as f:
            data = json.load(f)
        if isinstance(data, list):
            val_records.extend(data)
        else:
            val_records.append(data)
val_QA_df = pd.DataFrame(val_records)

# 결과 확인
print("Validation QA DataFrame shape:", val_QA_df.shape)
val_QA_df.head()
```

```python
import json
import pandas as pd

# 1) data_info에서 documentID 뽑아내기
val_QA_df['documentID'] = val_QA_df['data_info'].apply(lambda info: info['documentID'])

# 2) QA 리스트 하나당 한 행으로 펼치기
val_QA_df = val_QA_df.explode('QA').reset_index(drop=True)

# 3) question, answer 컬럼 분리
val_QA_df['question'] = val_QA_df['QA'].apply(lambda qa: qa['question'])
val_QA_df['answer']   = val_QA_df['QA'].apply(lambda qa: qa['answer'])
val_QA_df = val_QA_df.drop(columns=['QA'])

print("Validation QA columns:", val_QA_df.columns.tolist())

#    cleaned_val_df에는 '문단 번호', '한국어' 컬럼이 있어야 합니다
val_df = pd.merge(
    val_QA_df,
    val_clean[['문단 번호','한국어']],
    left_on='documentID',
    right_on='문단 번호',
    how='inner'
)
print("Validation set size:", val_df.shape)

# 5) JSONL 형식으로 저장
with open('QA_validation_dataset.jsonl', 'w', encoding='utf-8') as fout:
    for _, row in val_df.iterrows():
        context  = row['한국어'].replace('\n', ' ').strip()
        question = row['question'].replace('\n', ' ').strip()
        answer   = row['answer'].strip()

        rec = {
            "instruction": "한국어 글을 읽고 질문에 답하세요",
            "input":       f"【本文（한국語）】\n{context}\n\n【質問（日本語）】\n{question}",
            "output":      answer
        }
        fout.write(json.dumps(rec, ensure_ascii=False) + "\n")

print("✅ QA_validation_dataset.jsonl 생성 완료")
```
